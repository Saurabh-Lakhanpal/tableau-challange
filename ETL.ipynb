```python
#!/usr/bin/env python
# coding: utf-8

# Import necessary libraries
import os
import requests
import zipfile
import shutil

# List of desired CSV files
csv_files = [
    "JC-202207-citbike-tripdata.csv",
    "JC-202208-citibike-tripdata.csv",
    "JC-202209-citibike-tripdata.csv",
    "JC-202210-citibike-tripdata.csv",
    "JC-202211-citibike-tripdata.csv",
    "JC-202212-citibike-tripdata.csv",
    "JC-202301-citibike-tripdata.csv",
    "JC-202302-citibike-tripdata.csv",
    "JC-202303-citibike-tripdata.csv",
    "JC-202304-citibike-tripdata.csv",
    "JC-202305-citibike-tripdata.csv",
    "JC-202306-citibike-tripdata.csv",
    "JC-202307-citibike-tripdata.csv"
]

# URL prefix for the files
base_url = "https://s3.amazonaws.com/tripdata/"

# Directory to save the files
resources_dir = "Resources"

# Create the directory if it doesn't exist
os.makedirs(resources_dir, exist_ok=True)

# Download, unzip, delete ZIP files, and remove _MACOSX folders
for file_name in csv_files:
    zip_file_name = file_name + ".zip"
    file_url = base_url + zip_file_name
    zip_output_path = os.path.join(resources_dir, zip_file_name)

    print(f"Downloading {zip_file_name}...")
    response = requests.get(file_url)
    
    if response.status_code == 200:
        # Save the ZIP file
        with open(zip_output_path, "wb") as f:
            f.write(response.content)
        print(f"Saved {zip_file_name} to {resources_dir}")
        
        # Unzip the file
        print(f"Unzipping {zip_file_name}...")
        with zipfile.ZipFile(zip_output_path, "r") as zip_ref:
            zip_ref.extractall(resources_dir)
        print(f"Unzipped {zip_file_name} into {resources_dir}")
        
        # Delete the ZIP file
        os.remove(zip_output_path)
        print(f"Deleted {zip_file_name}")
        
        # Check for and remove the _MACOSX folder
        macosx_folder = os.path.join(resources_dir, "__MACOSX")
        if os.path.exists(macosx_folder):
            shutil.rmtree(macosx_folder)
            print("Deleted _MACOSX folder")
    else:
        print(f"Failed to download {zip_file_name}. HTTP Status Code: {response.status_code}")

print("All files processed.")
```
